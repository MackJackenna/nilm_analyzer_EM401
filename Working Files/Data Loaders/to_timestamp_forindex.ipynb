```python
import time
import calendar 
import datetime
from matplotlib import rcParams
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Read unformatted UKDALE mains data CSV into a pandas Dataframe.
df = pd.read_csv(r'...path...\nilm_analyzer_EM401\Large Data\Reactive Power Mains Data\h5_mains.csv', index_col=0, header=[0,1])
df.head()
```

```python
#Push the datetime64 index into a seperate column to perform conversion caluclation.
df = df.reset_index(inplace=False)
df.head()
```

```python
#Function that converts the datetime64 format from NILMTK to it's corresponding UNIX timestamp.
def to_timestamp(index):
    dt64 = np.datetime64('{}'.format(index))
    ts = (dt64 - np.datetime64('1970-01-01T00:00:00Z')) / np.timedelta64(1, 's')
    return ts
```

```python
#Applies to-timestamp function to the datetime64 column.
df['index'] = df['index'].apply(to_timestamp)
df.head()
```

```python
#Restablishes the the now UNIX column as the dataframes index.
df = df.set_index('index')
df.head()
```

```python
#Output to a CSV.
df.to_csv(r'...path...\nilm_analyzer_EM401\Large Data\Reactive Power Mains Data\h5_mains_formatted.csv')
```
