```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
RdfDW = pd.read_csv(r'...path...\nilm_analyzer_EM401\Working Files\Visualisation\Activations Data\UKDALE_dishwasher_activations.csv', index_col = 0)
RdfWM = pd.read_csv(r'...path...\nilm_analyzer_EM401\Working Files\Visualisation\Activations Data\UKDALE_washingmachine_activations.csv', index_col = 0)
```

```python
dfDW.sort_values(by = 'duration_in_minutes', ascending = True, inplace=True)
dfWM.sort_values(by = 'duration_in_minutes', ascending = True, inplace=True)
```

```python
douDW = dfDW['duration_in_minutes'].round(decimals=0).astype(int)
douWM = dfWM['duration_in_minutes'].round(decimals=0).astype(int)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (15,6), sharey= True)

fig.suptitle('UKDALE Combined Houses- Duration of Use ',fontsize = 20)
sns.histplot(douDW, bins='auto', kde=True, color='blue', edgecolor = 'red', ax = ax[0])
sns.histplot(douWM, bins= 20, kde=True, color='blue', edgecolor = 'red', ax= ax[1])

ax[0].set_xlabel('Duration [minutes]',fontsize = 12)
ax[1].set_xlabel('Duration [minutes]',fontsize = 12)
ax[0].set_ylabel('Number of Activations',fontsize = 12)
ax[1].set_ylabel('Activations',fontsize = 12)

ax[0].set_title('Dishwasher',fontsize = 16)
ax[1].set_title('Washing Machine',fontsize = 16)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)
plt.show()
```
